```python
# We use pivot data where each region is a column column has 3 values for NDVI mean for each year
# This can further we used for CP and avg Temp by pivoting data and having each region is a column column has 3 values for 
# cp and avg temp for each year

# libraries and data
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
 
# Make a data frame
df=pd.DataFrame({'x': german_draught_pivot_df['Year'], 
                 'C1': german_draught_pivot_df['1'], 
                 'C3': german_draught_pivot_df['3'], 
                 'C14': german_draught_pivot_df['14'], 
                 'C17': german_draught_pivot_df['17'], 
                 'C27': german_draught_pivot_df['27'], 
                 'C34': german_draught_pivot_df['34'],
                'C38': german_draught_pivot_df['38'],
                'C23': german_draught_pivot_df['23']})
 
# Initialize the figure
plt.style.use('seaborn-darkgrid')
 
# create a color palette
palette = plt.get_cmap('Set1')
 
# multiple line plot
num=0
for column in df.drop('x', axis=1):
    num+=1
 
    # Find the right spot on the plot
    plt.subplot(3,3,num)
    plt.tight_layout()
    # Plot the lineplot
    plt.plot(df['x'], df[column], marker='', color=palette(num), linewidth=1.9, alpha=0.9, label=column)
 
    # Add title
    plt.title(column, loc='left', fontsize=12, fontweight=0, color=palette(num) )
  
# general title
plt.suptitle("line plot showing cordinate wise trend in values of NDVI mean over the years", fontsize=13, fontweight=0, color='black', style='italic')
```
